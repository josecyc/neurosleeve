### Exploring the different hand gestures as spectrograms, with whole time period of 15 seconds

# Alpha 


```python
file_alpha['ch1'].shape
```

```python
nperseg=80
noverlap=75
frequencies, times, spec = signal.spectrogram(x=file_alpha['ch4'], fs=200, nperseg=nperseg, noverlap=noverlap, window='hann')
print('spec_shape', spec.shape) #'spec', spec)
print('frequencies_shape', frequencies.shape)
print('times_shape', times.shape)
#spec = np.log(spec)
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1 = plt.pcolormesh(times, frequencies, spec)
ax1 = plt.show
print(spec.shape)
frequencies, times, spec = signal.spectrogram(x=file_alpha['ch2'], fs=200, nperseg=nperseg, noverlap=noverlap, window='hann')
log_spec = np.log(spec)
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1 = plt.pcolormesh(times, frequencies, spec)
ax1 = plt.show
print(spec.shape)
frequencies, times, spec = signal.spectrogram(x=file_alpha['ch3'], fs=200, nperseg=nperseg, noverlap=noverlap, window='hann')
log_spec = np.log(spec)
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1 = plt.pcolormesh(times, frequencies, spec)
ax1 = plt.show
print(spec.shape)
frequencies, times, spec = signal.spectrogram(x=file_alpha['ch4'], fs=200, nperseg=nperseg, noverlap=noverlap, window='hann')
log_spec = np.log(spec)
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(211)
ax1 = plt.pcolormesh(times, frequencies, spec)
ax1 = plt.show
print(spec.shape)
```
